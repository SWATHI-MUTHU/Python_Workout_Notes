# Matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(2,3))
plt.scatter(x,y,c="#aa36f2")
plt.title("x vs y")
plt.xlabel("x")
plt.ylabel("y")
```

```python
#Dealing with cat and num?
x=['a','b','c','d','e']
y=np.random.rand(5)
```

```python
plt.figure(figsize=(2,3))
plt.bar(x,y)
plt.title("x vs y")
plt.xlabel("x")
plt.ylabel("y")
```
